```python
import json
with open('../log', 'r') as f:
    data = json.load(f)
```

```python
def build_series(name, data):
    for elem in data:
        yield elem[name]
```

```python
lr = list(build_series('lr', data))
loc = list(build_series('main/loss/conf', data))
conf = list(build_series('main/loss/loc', data))
epochs = list(build_series('epoch', data))
```

```python
import matplotlib.pyplot as plt
f, loss_plt = plt.subplots()
lr_plt = loss_plt.twinx()

loss_plt.scatter(epochs, conf, label='conf')
loss_plt.scatter(epochs, loc, label='loc')
loss_plt.legend(loc='center right')
loss_plt.set_yscale('log')
loss_plt.set_xlabel('epochs')

lr_plt.plot(epochs, lr)
lr_plt.set_ylabel('learning rate')
plt.tight_layout()
plt.show()
```
